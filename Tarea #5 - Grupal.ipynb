**Problema: "Gestión Sostenible de Residuos Electrónicos".**

Gestión de elementos registrados.

Reciclaje y desecho apropiado de componentes electrónicos.

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/EWASTE_06112023065410164.csv')
#Fuente: https://stats.oecd.org/Index.aspx?DataSetCode=EWASTE

print("Datos Completos:")
print(df)

print()
print("////////////////////////////////////////////")
print("////////////////////////////////////////////")
print()

conteoWaste = df['Waste category'].value_counts()
sumaValue = df.groupby('Waste category')['Value'].sum()

print("Conteo de ítems por categoría:")
print(conteoWaste)

print()
print("////////////////////////////////////////////")
print("////////////////////////////////////////////")
print()

print("Suma del peso(kg) de los items:")
print(sumaValue)

print()
print("////////////////////////////////////////////")
print("////////////////////////////////////////////")
print()

print("")
suma_por_pais_y_operaciones = df.groupby(['Country', 'Waste management operations'])['Value'].sum()
print(suma_por_pais_y_operaciones)
```
